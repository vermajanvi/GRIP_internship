#### The Above map looks untidy even though it can be zoomed in to view the Country in question. Hence in the next chart, I have used Folium's Marker Cluster to cluster these icons. This makes it visually pleasing and highly interactive

```python
m1 = folium.Map(location=[20, 0], tiles="CartoDB positron", zoom_start=2)
marker_cluster = MarkerCluster(
    name='clustered icons',
    overlay=True,
    control=False,
    icon_create_function=None
)
for i in range(0,len(terror_df_group)):
    marker=folium.Marker([terror_df_group.iloc[i]['latitude'],terror_df_group.iloc[i]['longitude']]) 
    popup='Group:{}<br>Country:{}'.format(terror_df_group.iloc[i]['Group'],
                                          terror_df_group.iloc[i]['Country'])
    folium.Popup(popup).add_to(marker)
    marker_cluster.add_child(marker)
marker_cluster.add_to(m1)
folium.TileLayer('openstreetmap').add_to(m1)
folium.TileLayer(
    tiles='https://api.mapbox.com/styles/v1/{id}/tiles/{z}/{x}/{y}?access_token={accessToken}',
    attr='Mapbox API © <a href="https://mapbox.com/about/maps">Mapbox</a>',
    name='Mapbox Bright',
    id='mapbox/bright-v9',
    accessToken='your-mapbox-api-access-token'
).add_to(m1)
folium.TileLayer('cartodbdark_matter').add_to(m1)
folium.TileLayer('stamentoner').add_to(m1)
folium.LayerControl().add_to(m1)
m1
```

```python
# Total Number of people killed in terror attack
killData = terror.loc[:,'Killed']
print('Number of people killed by terror attack:', int(sum(killData.dropna())))# drop the NaN values
```

```python
# Let's look at what types of attacks these deaths were made of.
attackData = terror.loc[:,'AttackType']
# attackData
typeKillData = pd.concat([attackData, killData], axis=1)
```

```python
typeKillData.head()
```

```python
typeKillFormatData = typeKillData.pivot_table(columns='AttackType', values='Killed', aggfunc='sum')
typeKillFormatData
```

```python
typeKillFormatData.info()
```

```python
labels = typeKillFormatData.columns.tolist() # convert line to list
transpoze = typeKillFormatData.T # transpoze
values = transpoze.values.tolist()
fig, ax = plt.subplots(figsize=(20, 15), subplot_kw=dict(aspect="equal"))
plt.pie(values, startangle=90, autopct='%.2f%%')
plt.title('Types of terrorist attacks that cause deaths')
plt.legend(labels, loc='upper right', bbox_to_anchor = (1.3, 0.9), fontsize=15) # location legend
plt.show()
```

#### <center>Armed assault and bombing/explosion are seen to be the cause of 77% of the deaths in these attacks. This rate is why these attacks are used so many times in terrorist actions. This is how dangerous weapons and explosives are to the world.</center>

# Iraq terrorism analysis

```python
# Filter the data for Iraq
Iraq = terror[terror.Country == 'Iraq']

# Get the top 5 cities in Iraq with the most terrorist attacks
Iraq_cities = Iraq.groupby(by='city', as_index=False).count().sort_values(by='Year', ascending=False).iloc[:5, :]
print("Top 5 cities in Iraq with the most terrorist attacks:")
print(Iraq_cities)
```

```python
# Calculate the proportion of suicide attacks in Iraq
Iraq_suicide_size = Iraq['Suicide'].sum() / len(Iraq)
print("\nProportion of suicide attacks in Iraq:")
print(Iraq_suicide_size)
```

```python
# Get the total number of deaths by year in Iraq
Iraq_year = Iraq.groupby(by='Year', as_index=False).sum().loc[:, ['Year', 'Killed']]
print("\nTotal number of deaths by year in Iraq:")
print(Iraq_year)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

# Plot 1 - Top 5 terrorism cities    
sns.barplot(x='Year', y='city', data=Iraq_cities, ci=None, ax=axs[0],palette='summer')
axs[0].set_title(f'Top 5 Iraq Cities With Most Terrorism Occurences')
axs[0].set_ylabel('City')
axs[0].set_xlabel('Victims')

# Plot 2 - Suicide Rate
center_circle = plt.Circle((0,0), 0.75, color='white')
axs[1].pie((Iraq_suicide_size, 1-Iraq_suicide_size), labels=['Suicide', 'Not Suicide'], colors=['crimson','green'] , autopct='%1.1f%%')
axs[1].add_artist(center_circle)
axs[1].set_title('Iraq Terrorism Suicide Rate')
axs[1].set_ylabel('Victims')

plt.suptitle('Terrorism Analysis in Iraq between 1970 and 2017', size=16)    
plt.subplots_adjust(top=0.90)
plt.show()
```

```python
plt.figure(figsize=(8, 6)) # Adjust the size of the figure

plt.plot(Iraq_year['Year'], Iraq_year['Killed'])
plt.xlabel('Year', fontsize=12) # Adjust the font size of the x-axis label
plt.ylabel('Number of Deaths', fontsize=12) # Adjust the font size of the y-axis label
plt.title('Number of Deaths by Year in Iraq', fontsize=14) # Adjust the font size of the title

plt.show()
```
